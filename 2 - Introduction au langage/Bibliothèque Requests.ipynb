```python
# Définir l'URL
url = "https://www.journalducameroun.com"

# Requête URL
reponse = requests.get(url)
contenu = reponse.text
contenu.count("chloroquine")

# Sauvegarder la page
with open("page_site.html","w",encoding="utf8") as file:
    file.write(contenu)
```

Petit exercice : 

Récupérer la page de trois quotidiens (lemonde, libération et lefigaro) et stocker dans un dictionnaire "url":"texte

```python
sites = [
         "https://www.lemonde.fr/",
         "https://www.liberation.fr/",
         "https://www.lefigaro.fr/",
        ]

for url in sites:
    req = requests.get(url)
```

```python
# Créer la structure des données
sites = {
         "https://www.lemonde.fr/":None,
         "https://www.liberation.fr/":None,
         "https://www.lefigaro.fr/":None,
        }

# Récupérer les informations sur internet
for url in sites:
    req = requests.get(url)
    
    # Associer le contenu textuel cmome valeur au dictionnaire
    sites[url] = req.text

# Sauvegarder sur un fichier
with open("sauvegarde.txt","w",encoding="utf8") as f:
    f.write(str(sites))
```
